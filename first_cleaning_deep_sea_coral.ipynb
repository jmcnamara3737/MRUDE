```python
import pandas as pd
```

```python
data = pd.read_csv("/Users/alyssaesca/Downloads/deep_sea_corals.csv")

data.head()
```

```python
# dropping rows with missing values
data = data.dropna()
```

```python
# reformatting dates to yyyy-mm-dd
data['Date_Collected'] = pd.to_datetime(data['ObservationDate'], errors='coerce').dt.strftime('%Y-%m-%d')
```

```python
data.info()
data.describe(include='all')
```

```python
# dropping columns that are not relevant
columns_to_drop = ['CatalogNumber', 'EventID', 'SampleID']  
data = data.drop(columns=columns_to_drop)
```

```python
data.to_csv('/Users/alyssaesca/Downloads/processed_deep_sea_corals.csv', index=False)
```
